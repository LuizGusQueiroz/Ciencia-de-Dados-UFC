## OUTRA FORMA DE APRESENTAR AS MEDIDAS

### GUARDANDO AS MEDIDAS EM UM DATAFRAME

```python
medidas = {
    'Mean':[],
    'Mode':[],
    'Median':[],
    'Max Value':[],
    'Min Value':[],
    'Amplitude':[],
    'Intervalo Interquartil':[],
    'Variance':[],
    'Desvio-Padrão':[],
    'Coeficiente De Variação':[]
}
```

```python
medidas['Mean'] = [
    round(df["Preço"].mean(), 2),
    round(df["Ano Fabricação"].mean()),
    round(df["Ano Modelo"].mean()),
    df["Quilometragem"].mean()]

medidas['Mode'] = [
    df["Preço"].mode()[0],
    df["Ano Fabricação"].mode()[0],
    df["Ano Modelo"].mode()[0],
    df["Quilometragem"].mode()[0]
]
medidas['Median'] = [
    df["Preço"].median(),
    round(df["Ano Fabricação"].median()),
    round(df["Ano Modelo"].median()),
    round(df["Quilometragem"].median(), 2)
]
medidas['Max Value'] = [
    df["Preço"].max(),
    df["Ano Fabricação"].max(),
    df["Ano Modelo"].max(),
    df["Quilometragem"].max()
]
medidas['Min Value'] = [
    df["Preço"].min(),
    df["Ano Fabricação"].min(),
    df["Ano Modelo"].min(),
    df["Quilometragem"].min()
]
medidas['Amplitude'] = [
    df["Preço"].max()-df["Preço"].min(),
    df["Ano Fabricação"].max()-df["Ano Fabricação"].min(),
    df["Ano Modelo"].max()-df["Ano Modelo"].min(),
    df["Quilometragem"].max()-df["Quilometragem"].min()
]
medidas['Intervalo Interquartil'] = [
    int(df["Preço"].quantile(0.75)-df["Preço"].quantile(0.25)),
    int(df["Ano Fabricação"].quantile(0.75)-df["Ano Fabricação"].quantile(0.25)),
    int(df["Ano Modelo"].quantile(0.75)-df["Ano Modelo"].quantile(0.25)),
    int(df["Quilometragem"].quantile(0.75)-df["Quilometragem"].quantile(0.25))
]
medidas['Variance'] = [
    round(df["Preço"].var(), 2),
    round(df["Ano Fabricação"].var(), 2),
    round(df["Ano Modelo"].var(), 2),
    round(df["Quilometragem"].var(), 2)
]
medidas['Desvio-Padrão'] = [
    round(df["Preço"].std(), 2),
    round(df["Ano Fabricação"].std(), 2),
    round(df["Ano Modelo"].std(), 2),
    round(df["Quilometragem"].std(), 2)
]
medidas['Coeficiente De Variação'] = [
    round((df["Preço"].std()/df["Preço"].mean())*100, 2),
    round((df["Ano Fabricação"].std()/df["Ano Fabricação"].mean())*100, 2),
    round((df["Ano Modelo"].std()/df["Ano Modelo"].mean())*100, 2),
    round((df["Quilometragem"].std()/df["Quilometragem"].mean())*100, 2)
]
# Criando o DataFrame
dfMedidas = pd.DataFrame(medidas)
# Renomeando as linhas
linhas = ['Preço', 'Ano Fabr', 'Ano Model', 'KM']
dfMedidas['nomes'] = linhas
dfMedidas = dfMedidas.set_index('nomes')
```

```python
# Cria um dataFrame com cada marca e sua frequência
contagem = df['Marca'].value_counts().reset_index()
# Nomeando as colunas
contagem.columns = ['Marca', 'Frequência']
# Separando as 25 mais Frequentes
top25 = contagem.head(25).sort_values(by='Frequência', ascending=True)
# Criando gráfico de barras horizontais
plt.barh(top25['Marca'], top25['Frequência'])
plt.xlabel('Número de carros')
plt.ylabel('Marca')
plt.title('25 Marcas Mais Frequentes')
plt.show()
```

```python
# Cria um dataFrame com cada motor e sua frequência
contagem = df['Motor'].value_counts().reset_index()
# Nomeando as colunas
contagem.columns = ['Motor', 'Frequência']
# Separando as 25 mais Frequentes
top25 = contagem.head(25).sort_values(by='Frequência', ascending=True)
# Criando gráfico de barras horizontais
plt.barh(top25['Motor'], top25['Frequência'])
plt.xlabel('Número de carros')
plt.ylabel('Potência do motor')
plt.title('25 Potências de motor mais frequentes')
plt.show()
```

```python
# Limpando o DataFrame tabFreq
tabFreq = pd.DataFrame()
# Definindo os intervalos
intervalos = [122, 50000, 100000, 150000, 200000, 250000, 300000, 350000, 400000, 500000, 1000000, 2000000, 202900000]
# Definindo os rótulos
rotulos = ['120-49999'    , '50000-100000'    , '100000-150000',
           '150000-200000', '200000-250000'  , '250000-300000',
           '300000-350000', '350000-400000'  , '400000-500000',
           '500000-1000000', '1000000-2000000', '2000000-202900000']
# Criando um dataFrame auxiliar para calcular a contagem dos preços nos intervalos
dfAux = pd.DataFrame()
# Guardando os valores agrupando pelos intervalos aplicados na coluna Preço
dfAux['Intervalo'] = pd.cut(df['Preço'], bins=intervalos, labels=rotulos, include_lowest=True)
# Calcular a frequência dos valores em cada intervalo
tabFreq = dfAux['Intervalo'].value_counts().reset_index()
# Renomeando as colunas
tabFreq.columns = ['Intervalo', 'Frequência']
# Ajuste para os valores poderem ser ordenados pelos intervalos
tabFreq['Intervalo'] = pd.Categorical(tabFreq['Intervalo'], categories=rotulos, ordered=True)
# Ordenando as linhas pelos intervalos
tabFreq = tabFreq.sort_values('Intervalo')
# Criando coluna da Frequência relativa,
#que é a frequência dividida pelo tamanho do DataFrame
tabFreq['Freq Rel'] = tabFreq['Frequência']/(df.shape)[0]
# Criando coluna da Frequência relativa,
#que é a Frequência relativa vezes 100
tabFreq['Freq Perc'] = tabFreq['Freq Rel'] * 100
# Criando coluna de Frequência acumulada
tabFreq['Freq Acumulada'] = tabFreq['Frequência'].cumsum()
# Criando função lambda para transformar a String do intervalo
#em dois valores inteiros para poder calcular sua diferença
# Criando coluna Densidade,
#que é a Frequência Relativa dividida pela amplitude.
# A Densidade foi multiplicada por 10^6 para facilitar a visualização dos valores
amplitude = lambda x: int(x.split('-')[1]) - int(x.split('-')[0])
tabFreq['Densidade * 10^6'] = 10**6 * (tabFreq['Freq Rel'] / tabFreq['Intervalo'].apply(amplitude))
# Criando função lambda para transformar a String do intervalo
#em dois valores inteiros para poder calcular a sua média
pontoMedio = lambda x: (int(x.split('-')[0])+int(x.split('-')[1]))/2
# criando coluna Ponto Médio,
#que é a média do intervalo
tabFreq['Ponto Médio'] = tabFreq['Intervalo'].apply(pontoMedio)
# Atualizando os índices
tabFreq = tabFreq.reset_index(drop=True)
# Adicionando título
tabFreq = tabFreq.style.set_caption('Tabela de frequência do Preço dos carros')
```

```python
# Limpando o DataFrame tabFreq
tabFreq = pd.DataFrame()
# Definindo os intervalos
intervalos = [0, 50000, 100000, 150000, 200000, 250000, 500000, 2147483647]
# Definindo os rótulos
rotulos = ['0-50000'    , '50000-100000'    , '100000-150000',
           '150000-200000', '200000-250000'  , '250000-500000',
           '500000-21474836470']
# Criando um dataFrame auxiliar para calcular a contagem dos preços nos intervalos
dfAux = pd.DataFrame()
# Guardando os valores agrupando pelos intervalos aplicados na coluna Preço
dfAux['Intervalo'] = pd.cut(df['Quilometragem'], bins=intervalos, labels=rotulos, include_lowest=True)
# Calcular a frequência dos valores em cada intervalo
tabFreq = dfAux['Intervalo'].value_counts().reset_index()
# Renomeando as colunas
tabFreq.columns = ['Intervalo', 'Frequência']
# Ajuste para os valores poderem ser ordenados pelos intervalos
tabFreq['Intervalo'] = pd.Categorical(tabFreq['Intervalo'], categories=rotulos, ordered=True)
# Ordenando as linhas pelos intervalos
tabFreq = tabFreq.sort_values('Intervalo')
# Criando coluna da Frequência relativa,
#que é a frequência dividida pelo tamanho do DataFrame
tabFreq['Freq Rel'] = tabFreq['Frequência']/(df.shape)[0]
# Criando coluna da Frequência relativa,
#que é a Frequência relativa vezes 100
tabFreq['Freq Perc'] = tabFreq['Freq Rel'] * 100
# Criando coluna de Frequência acumulada
tabFreq['Freq Acumulada'] = tabFreq['Frequência'].cumsum()
# Criando função lambda para transformar a String do intervalo
#em dois valores inteiros para poder calcular sua diferença
# Criando coluna Densidade,
#que é a Frequência Relativa dividida pela amplitude.
# A Densidade foi multiplicada por 10^6 para facilitar a visualização dos valores
amplitude = lambda x: int(x.split('-')[1]) - int(x.split('-')[0])
tabFreq['Densidade * 10^6'] = 10**6 * (tabFreq['Freq Rel'] / tabFreq['Intervalo'].apply(amplitude))
# Criando função lambda para transformar a String do intervalo
#em dois valores inteiros para poder calcular a sua média
pontoMedio = lambda x: int((int(x.split('-')[0])+int(x.split('-')[1]))/2)
# criando coluna Ponto Médio,
#que é a média do intervalo
tabFreq['Ponto Médio'] = tabFreq['Intervalo'].apply(pontoMedio)
# Atualizando os índices
tabFreq = tabFreq.reset_index(drop=True)
# Adicionando título
tabFreq = tabFreq.style.set_caption('Tabela de frequência da Quilometragem dos carros')
```
